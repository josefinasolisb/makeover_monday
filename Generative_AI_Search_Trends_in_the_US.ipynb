```python
from google.colab import drive
drive.mount('/content/drive')
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
filename = '/content/drive/MyDrive/Colab Notebooks/Coding Dojo/S4_ExplanatoryDataViz/percentage_of_searches_by_US_subregion.csv'
df = pd.read_csv(filename)
df.head()
```

```python
df.info()
```

```python
df = df.rename(columns={
    'Midjourney: (1/1/22 - 2/16/24)': 'Midjourney',
    'Stable Diffusion: (1/1/22 - 2/16/24)': 'Stable Diffusion',
    'DALL E: (1/1/22 - 2/16/24)': 'DALL E'
})
df
```

```python
columns = ['Midjourney', 'Stable Diffusion', 'DALL E']
colors = ['royalblue', 'indianred', 'gold']
```

```python
df[columns] = df[columns].replace('%', '', regex=True).astype(int)
```

```python
plt.figure(figsize=(15, 6))

for i, column in enumerate(columns):
    sns.barplot(x='Region', y=column, data=df, label=column, color=colors[i])

plt.title('Generative AI Search Trends in the United States', fontsize=22)
plt.xlabel('Region', fontsize=15)
plt.ylabel('Percentage', fontsize=15)
plt.xticks(rotation=90)

plt.legend()

plt.show()
```
